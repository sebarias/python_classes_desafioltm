### <font color='green'> SUBSET </font>

<p style="font-family: Arial; font-size:1.4em;color:grey;"> 

     * Extraer subconjuntos especificos del dataframe.
    
     * Para acceder a alguna observación específica (ya sea fila o columna), se dene hacer mediante a notación de loc ó iloc.
         
      * loc: Localiza una observación específica mediante el nombre de una columna y su posición. df[fila,columna] 
     
     * También es posible acceder a la fila completa, incuyendo todas las columnas. Esto se logra mediante el operador : (slice). Éste indica que se está considerando todos los elementos de una lista o serie (los nombres de las columnas también son una Serie). df.loc[fila,:]
     
     * También es posible hacer un subset de datos seleccionando todas las filas y un rango específico de
columnas.  df.loc[:,['variable1','variable2',...,'variablen']]

  *  En caso de que se requiera seleccionar "desde" una columna "hasta" otra columna (incluyendo todas las
columnas intermedia)s, también es posible el uso de slice : . Como slice va a seleccionar todo, sólo sirve si
el rango de columnas que se desea es contínuo. df.loc[:, 'variable1':'variablen']

</p>


```python
# Localizar una observación específica de la columna school y la posicion 9.
df.loc[8, 'school']
```

```python
# extraer todos los registros de los campos en la novena observación.
df.loc[8,:]
```

### <font color='green'> ITERACION EN UN DATAFRAME </font>

<p style="font-family: Arial; font-size:1.4em;color:grey;"> 

     * Puede ser necesario para el flujo de trabajo, es iterar fila a fila y evaluar cada una para algún caso específico.
    
     * ¿Cómo iterar las filas o registros del DataFrame?
     
     * Existe más de una forma de poder recorrer los registros, entre estos están los métodos df.iterrows() , df.iteritems()  
              
</p>

```python
# iterando un dataframe (solo se obtiene los nombres de las columnas.)

for i in df:
    print(i)
```

### <font color='green'> df.iteritems() </font>

<p style="font-family: Arial; font-size:1.4em;color:grey;"> 

     * Iteración es posible recorrer todos los elementos de un DataFrame por columna. Esto significa que se va a estar recorriendo todos los campos de una base de datos a la vez. 
     
     * De similar manera a cuando se utilizaba enumerate , necesita tener dos iterables declarados. El primer iterable responde al nombre de la columna, mientras que el segundo devuelve los valores asociados a cada columna en forma de Serie . 
     
     * Es posible concatenar acciones que operen a nivel de serie dentro del loop.
              
</p>

```python
for colname, serie in df.iteritems():
    print(colname)
```

```python
for colname, serie in df.iteritems():
    print(colname)
    print(serie.unique())
    break
    
```

```python
for colname, serie in df.iteritems():
    if colname =='region':
        print(serie.unique())
        print(type(colname))
        print(type(serie))
```

```python
# inspeccionemos si todos los datos corresponden al tipo numérico o no, se hará uso de uns función de pandas : pd.api.types.is_numeric_dtype .

for colname, serie in df.iteritems():
    tmp = pd.api.types.is_numeric_dtype(serie)
    print("{} es {}".format(colname, tmp))
```

### <font color='green'> df.iterrrows() </font>

<p style="font-family: Arial; font-size:1.4em;color:grey;"> 

     * iteracion permite recorrer todos los elementos de un DataFrame por fila. Esto significa que se va a estar recorriendo todas las columnas, por registro, de una base de datos a la vez.
    
                 
</p>

```python
for index, Serie in df.iterrows():
    print(index)
    print(Serie)
    break
```

```python
for index, row_serie in df.iterrows():
    if index == 1:
        print(row_serie)
        print(type(row_serie))
```
